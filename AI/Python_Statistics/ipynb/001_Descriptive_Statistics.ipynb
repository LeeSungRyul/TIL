# Descriptive Statistics

```
import warnings
warnings.filterwarnings('ignore')
```

# Load 'tips.csv' Data

```
import seaborn as sns

DF = sns.load_dataset('tips')
```

# I. pandas

> ## 1) DataFrame Information

```
DF.info()
```

```
DF.head()
```

> ## 2) .describe()

```
DF.tip.describe()
```

> ## 3) .sum()

```
%precision 5  # 소수점 5자리까지

DF.tip.sum()
```

> ## 4) .mean()

```
DF.tip.mean()
```

> ## 5) .min()

```
DF.tip.min()
```

> ## 6) .quantile(q = 0.25)

```
DF.tip.quantile(q = 0.25) # 25%
```

> ## 7) .median()

```
DF.tip.median() # q = 0.5와 동일
```

> ## 8) .quantile(q = 0.75)

```
DF.tip.quantile(q = 0.75) # 75%
```

> ## 9) .max()

```
DF.tip.max()
```

> ## 10) .var(ddof = 0)

- Dynamic Degrees Of Freedom(default = 1)

```
DF.tip.var(ddof = 0)  # 분산. ddof: 불편분산
```

> ## 11) .std(ddof = 0)

```
DF.tip.std(ddof = 0)  # 표준편차
```

# III. scipy

```
import scipy as sp
```

> ## 1) .sum()

```
sp.sum(DF['tip'])
```

> ## 2) .mean()

```
sp.mean(DF['tip'])
```

> ## 3) .amin()

```
sp.amin(DF['tip'])
```

> ## 4) .median()

```
sp.median(DF['tip'])
```

> ## 5) .amax()

```
sp.amax(DF['tip'])
```

> ## 6) .var(ddof = 0)

- defualt = 0

```
sp.var(DF['tip'], ddof = 0)
```

> ## 7) .std(ddof = 0)

```
sp.std(DF['tip'], ddof = 0)
```

> ## 8) .stats.mode()

```
sp.stats.mode(DF['tip'])
```
