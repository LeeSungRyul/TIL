## 7.2 부품 찾기

```python
def binary_search(array, target, start, end):
  while start <= end:   # < 로 하면 마지막에 9 출력 안 됨
    mid = (start + end) // 2
    if array[mid] == target:
      return mid
    elif array[mid] > target:
      end = mid - 1
    else:
      start = mid + 1
    print(f'mid: {mid}, target: {target}')
  return None
```

```python
N = int(input())
market = list(map(int, input().split()))
M = int(input())
targets = list(map(int, input().split()))

market.sort()   # 탐색 전 정렬 필요3


for target in targets:
  result = binary_search(market, target, 0, N - 1)

  if result != None:
    print(f'{target}: yes')
  else:
    print(f'{target}: no')
```

## 7.3 떡볶이 떡 만들기

```python
N, M = map(int, input().split())    # N: 떡의 개수, M: 요청한 떡의 길이
dduks = list(map(int, input().split()))

```

```python
def find_len(array, target, start, end):
  ans = 0

  while start <= end:
    mid = (start + end) // 2
    total = 0
    for i in array:
      if i > mid:
        total += i - mid

    if total < target:
      end = mid - 1
    else:
      ans = mid
      start = mid + 1

  return ans
```

```python
n, m = 4, 6
dduks = [19, 15, 10, 17]

result = find_len(dduks, m, 0, max(dduks))
print(result)
```
