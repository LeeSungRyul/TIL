# 3.2 큰 수의 법칙

```python
n, m, k = map(int, input().split(' '))
data = list(map(int, input().split(' ')))

data.sort(reverse = True)
max_first = data[0]
max_second = data[1]

ans = 0

while True:
  for i in range(k):
    if m == 0:
      break
    ans += max_first
    m -= 1
  
  if m == 0:
    break

  ans += max_second
  m -= 1

print(ans)
```

# 3.3 숫자 카드 게임

```python
n, m = map(int, input().split(' '))
data = []
for i in range(n):
  temp = list(map(int, input().split(' ')))
  data.append(temp)

ans = 0

for i in range(n):
  temp = min(data[i])
  if temp > ans:
    ans = temp

print(ans)  
```

# 3.4 1이 될 때까지

```python
n, k = map(int, input().split(' '))

ans = n % k
n -= ans
while n > 1:
  n //= k
  ans += 1
  
print(ans)
```
