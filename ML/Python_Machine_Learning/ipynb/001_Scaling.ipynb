# I. 실습 데이터

> ## 1) seaborn 'mpg' Data Set

```
import seaborn as sns

DF = sns.load_dataset('mpg')
```

- 자동차 연비(mpg) 데이터
  - mpg : miles per gallon
  - 1 mile : 1.6 km
  - 1 gallon : 3.78 L

```
DF.info()
```

```
DF.head() # train, test split X
```

- X, y Data

```
X = DF[['weight']]  # DF type
y = DF['mpg']
```

> ## 2) Without Scailing

- X 값 확인

```
X[:5]
```

```
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (9, 6))
sns.regplot(x = X, y = y)
plt.xlabel('weight_Without_Scailing')
plt.show()
```

> ## 3) With Normalization

- sklearn Package

```
from sklearn.preprocessing import MinMaxScaler

scaler1 = MinMaxScaler()
X_Norm = scaler1.fit_transform(X)
```

- 정규화된 X 값 확인

```
X_Norm[:5]
```

- 정규화된 X 값 모델 생성

```
fig = plt.figure(figsize = (9, 6))
sns.regplot(x = X_Norm, y = y)
plt.xlabel('weight_With_Normalization')
plt.show()
```

> ## 4) With Standardization

- sklearn Package

```
from sklearn.preprocessing import StandardScaler

scaler2 = StandardScaler()
X_Stan = scaler2.fit_transform(X)
```

- 표준화된 X 값 확인

```
X_Stan[:5]
```

- 표준화된 X 값 모델 생성

```
fig = plt.figure(figsize = (9, 6))
sns.regplot(x = X_Stan, y = y)
plt.xlabel('weight_With_Standardization')
plt.show()
```
